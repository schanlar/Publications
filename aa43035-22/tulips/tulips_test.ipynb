```python
import matplotlib.pyplot as plt
import mesaPlot as mp
import tulips
from IPython.display import Video

# Interactive matplotlib plotting for jupyter lab
%matplotlib inline
```

```python
EXAMPLE_DIR = "/Volumes/SavvasExternal/aifa_models/representatives/2.5000_0.0200_0.0000/LOGS/"

m0p25 = mp.MESA() # Create MESA object
m0p25.loadHistory(filename_in=EXAMPLE_DIR + 'history.data')
```

```python
star_age = m0p25.hist.star_age # Access star age
log_R = m0p25.hist.log_R # Access star log radius
```
