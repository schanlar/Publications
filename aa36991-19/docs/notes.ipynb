### Initial models

All of our helium star models contain the hydrogen-burning ashes of a $20\,{\rm M}_{\odot}$ ZAMS model. The star was evolved from its pre-main sequence contraction phase up until a $3.5\,{\rm M}_{\odot}$ was formed. The outer H and He layers were then removed using a constant mass loss rate of $\log [ \dot{M} / ({\rm M}_{\odot}{\rm yr}^{-1})]  = -3$. Our grid contains a total of 42 models, covering the mass range between 1.5 and $3.5\,{\rm M}_{\odot}$  with a step of $\Delta M = 0.1\,{\rm M}_{\odot}$ and metallicities of $Z=0.02$ and $0.001$. For all stellar models, the initial nuclear composition is constant throughout the star. The dominant species by mass fraction are $^4{\rm He}$:0.998,0.997 and $^{14}{\rm N}$ for the solar and sub-solar metallicity models respectively. 





### Numerics and Physics

MESA evaluates the stellar structure and composition equations simultaneously, using adaptive mesh and timestep refinement. The surface boundary conditions are set. We use the ''mesa'' equation-of-state 
All our models typically have between 1500 and 2500 cells. 

Nucleosynthesis is handled using a reaction network that tracks the evolution of the following 43 isotopes:
$n$,$^{1,2}$H, $^{3-4}$He, $^{7}$Li, $^{7}$Be, $^{8}$B, $^{12,13}$C, $^{13-15}$N,$^{16-18,20}$O,$^{12,20,23}$F,
$^{20-25}$Ne,$^{21-25}$Na, $^{24-26}$Mg, $^{24-26}$Mg, $^{25-27}$Al, $^{27,28}$Si, $^{30,31}$S, $^{56}$Fe, $^{58}$Ni. 
During the final stages of burning, we employ two slighly different reaction networks; one that tracks Urca and $e^{-}-$capture reactions, and one that does not. 

Convective stability was evaluated using the Ledoux criterion. We handle convective mixing using standard MLT theory, adopting a mixing length of $\alpha = 2.0$. Thermohaline mixing is evaluated using the perscription of Brown Garaud & Stellmach (2013)

#### Core evolution  and runaway 


In our MESA models the carbon flame always reaches the center. In other words, there is no hybrid CONe core formed like in the models discussed in some of these papers.  Rather, the core composition look very similar to the more recent sAGB stars of Schwab and Rocha (2018; https://arxiv.org/abs/1810.13002). 

The flame propagation and subsequent carbon distribution depends on the efficiency of thermohaline and boundary  mixing (similar to https://arxiv.org/abs/1407.0248). 
When thermohaline is not included, the mass of the residual carbon is generally smaller and the evolution looks very similar to the models recently published by Woosley (see https://arxiv.org/abs/1901.00215). Nevertheless, carbon and subsequently oxygen still ignite for some of them. Rotation may also play a role but apparently it has a lesser effect. Overall, regarding the amount of residual carbon, These models seem to be on the conservative side. 


The subsequent evolution depends on the competition between heating and cooling processes during contraction. As can be seen in the attached plot, all models undergo a carbon-simmering phase before any urca reaction takes place. Depending on the amount of residual carbon, some models avoid the critical density for Na23<-->Ne23 which seems to be the most efficient Urca process. 

For instance, the 2.3 Msol, Z=0.001 model has a final mass of 1.36 Msol and reaches central oxygen ingition at log rho~9.1. For this model, core degeneracy was lifted shortly before MESA crashed due to EoS issues.  

Other models reach higher densities and O ignites centraly due to 24Mg electron captures. However, all models with masses below 2.3 Msol avoid Ne20 ->Fe20 captures. 

Now the question is if the thermonuclear runaway destroys these stars completely. leading to type Ia SNe or if a NS forms instead. I guess the fact that some models are sub-Chandrasekhar and their central density reduces after O-ignition is encouraging. 

There was a recent paper by Jones et al. (https://arxiv.org/abs/1812.08230) looking at e-capture explosions. Apparently, their lower-density models which still have log rho_c > 9.9 leave behind a an ONeFe WD. 


For fun, I run a small pop-synthesis simulation with BSE to track the number of He cores with masses between 2.0 and 2.3 resulting from a single burst of star formation. The resulting "rates" are not too far off the observed type Ia delay time distribution. It should be interesting to do this properly.
